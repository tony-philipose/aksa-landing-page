```python
#https://www.dataquest.io/blog/pandas-pivot-table/
import numpy as np
import matplotlib.pyplot as pp
import pandas as pd
import seaborn as sb
```

```python
data = pd.read_excel("/home/local/ASSYST-COC/tony.p/Downloads/Financial Sample.xlsx")
data.head()
```

```python
pd.pivot_table(data,index=["Segment","Year"],values=["Profit"],aggfunc=[np.sum])
```

```python
val = pd.pivot_table(data,index=["Country","Year","Segment"],
               values=["Profit","Manufacturing Price","Units Sold"])
```

```python
val
```

```python
sb.barplot(x='Country', y='Year', hue='Profit', data=val)
```

```python
val.plot(kind='bar')
```

```python
val1 = pd.pivot_table(data,index=["Country","Year"],
               values=["Profit"])
val1
```

```python
val1.plot(kind='bar')
```

```python
d2 = pd.pivot_table(data,index=["Country","Year"],columns=["Segment"],values=["Profit"])
```

```python
d2.plot(kind='line')
```

```python
d2.plot(kind='bar')
```

```python
d2.plot(kind='Area')
```

```python
d2.plot(kind='Histogram')
```

```python
pd.pivot_table(data,index=["Segment","Year"],values=["Profit"],aggfunc=[np.sum]).plot(kind='bar')
```

```python
pd.pivot_table(data,index=["Segment","Year"],values=["Profit"],aggfunc=[np.sum]).plot(kind='line')
```

```python
pd.pivot_table(data,index=["Segment","Year"],values=["Profit"],aggfunc=[np.sum]).plot(kind='Area')
```
