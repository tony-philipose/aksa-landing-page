```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('/home/local/ASSYST-COC/tony.p/Downloads/f1000.xlsx')
data.head()
```

```python
data["Country"].head()
```

```python
data["Country"] = data["Country"].astype("category")
```

```python
data["Country"].cat.set_categories(["United States","Great Britain","France"],inplace=True)
```

```python
pd.pivot_table(data,index=["First Name","Last Name","Gender","Date"])
```

```python
pd.pivot_table(data,index=["Date","First Name","Last Name","Gender",])
```

```python
pd.pivot_table(data,index=["Date","First Name","Last Name"],values=["Id"])
```

```python
pd.pivot_table(data,index=["Date","First Name"],values=["Age"])
```

```python
pd.pivot_table(data,index=["Date","First Name"],values=["Age"],aggfunc=[np.mean])
```

```python
pd.pivot_table(data,index=["Date","First Name"],values=["Age"],aggfunc=[np.sum])
```

```python
pd.pivot_table(data,index=["Date","First Name"],values=["Age"],
               columns=["Country"],aggfunc=[np.sum])
```

```python
tm = data.groupby('Country').groups
tm = data.groupby('Country')
for row,group in tm:
    print (row)
    print (group)
```

```python
pd.pivot_table(data,index=["Date","First Name"],values=["Age"],
               columns=["Country"])
```

```python
pd.pivot_table(data,index=["Date","First Name"],values=["Age"],
               columns=["Country"])
```

```python
pd.pivot_table(data,index=["Gender","First Name"],values=["Age"],
               columns=["Date"],fill_value=0)
```

```python
pd.pivot_table(data,index=["Gender","First Name"],values=["Age","Date"],
               columns=["Country"],fill_value=0)
```

```python
pd.pivot_table(data,index=["Gender","First Name"],values=["Age","Date"],
               columns=["Country"],aggfunc=[np.sum],fill_value=0)
```
