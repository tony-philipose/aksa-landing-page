```python
#refer: http://pbpython.com/pandas-pivot-table-explained.html
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("/home/local/ASSYST-COC/tony.p/Downloads/Financial Sample.xlsx")
data.head()
```

```python
pd.pivot_table(data,index=["Segment"])
```

```python
pd.pivot_table(data,index=["Country","Segment","Month Name"])
```

```python
pd.pivot_table(data,index=["Country","Segment","Year"])
```

```python
pd.pivot_table(data,index=["Country","Segment","Year"],values=["Manufacturing Price","Sale Price","Profit","Discounts"])
```

```python
pd.pivot_table(data,index=["Segment","Year"],values=["Profit"],aggfunc=[np.sum],fill_value=0).head()
```

```python
pd.pivot_table(data,index=["Country","Year"],values=["Profit"],aggfunc=[np.sum,np.mean,len],fill_value=0).head()
```

```python
pd.pivot_table(data,index=["Country","Year"],values=["Profit"],
               columns=["Segment"],aggfunc=[np.sum,len])
```

```python
pd.pivot_table(data,index=["Country","Year"],values=["Profit","Manufacturing Price"],
               columns=["Segment"],aggfunc=[np.sum])
```

```python
pd.pivot_table(data,index=["Country","Year","Segment"],
               values=["Profit","Manufacturing Price"],
               aggfunc=[np.sum,np.mean],fill_value=0,margins=True)
```

```python
pd.pivot_table(data,index=["Country","Year","Segment"],
               values=["Profit","Manufacturing Price","Units Sold"])
```

```python
pd.pivot_table(data,index=["Country","Year"],columns=["Segment"],values=["Units Sold","Profit"],
               aggfunc={"Units Sold":np.sum,"Profit":np.sum},fill_value=0)
```

```python
pd.pivot_table(data,index=["Country","Year"],columns=["Segment"],values=["Units Sold","Profit"],
               aggfunc={"Units Sold":np.sum,"Profit":[np.sum,np.mean]},fill_value=0)
```
